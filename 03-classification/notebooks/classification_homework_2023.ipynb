```python
# Make price binary
# Now we need to turn the price variable from numeric into a binary format.
# Let's create a variable above_average which is 1 if the price is above its mean value and 0 otherwise.

price_mean_val = feature_renamed_df.price.mean()
feature_renamed_df['above_average'] = np.where(feature_renamed_df['price']>price_mean_val, 1,0)
```

```python
feature_renamed_df = feature_renamed_df.drop(['price'], axis=1)
```
